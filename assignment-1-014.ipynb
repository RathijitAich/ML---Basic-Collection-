```python
release_counts = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(12,6))
plt.plot(release_counts.index, release_counts.values, marker='o', color='green')
plt.title('Number of Titles Released Each Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.grid(True)
plt.show()

# this works better as the x axis is now sorted
```

```python
top_countries = df['country'].value_counts().head(20)

plt.figure(figsize=(10,6))
plt.barh(top_countries.index, top_countries.values, color='tomato')
plt.title('Top 10 Countries by Number of Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.grid(axis='x', linestyle='--')
plt.show()
```

```python
type_counts = df['type'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(type_counts.values, labels=type_counts.index, autopct='%1.1f%%', colors=['gold', 'red'])
plt.title('Proportion of Movies vs TV Shows')
plt.show()
```

```python
from collections import Counter

# Split multi-genres and count
all_genres = ','.join(df['listed_in'].dropna()).split(', ')
genre_counts = Counter(all_genres)

top_genres = dict(Counter(genre_counts).most_common(10))

plt.figure(figsize=(10,6))
plt.bar(top_genres.keys(), top_genres.values(), color='mediumpurple')
plt.title('Top 10 Genres on Netflix')
plt.xticks(rotation=45)
plt.ylabel('Number of Titles')
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
#Seaborn Implementation


plt.figure(figsize=(10,6))
sns.countplot(data=df, x='rating', order=df['rating'].value_counts().index[:10])
plt.xticks(rotation=45)
plt.title('Top Ratings on Netflix')
plt.show()

```

```python
yearly = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(12,6))
sns.lineplot(x=yearly.index, y=yearly.values)
plt.title('Number of Titles Released Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()
```

```python
df['type_encoded'] = df['type'].map({'Movie': 1, 'TV Show': 0})
corr = df[['release_year', 'type_encoded']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Extract number from "duration" (e.g., "90 min")
df['duration_mins'] = df['duration'].str.extract('(\d+)').astype(float)

sns.violinplot(data=df, x='type', y='duration_mins')
plt.title('Distribution of Duration by Type')
plt.show()
```

```python
sns.boxplot(data=df, x='type', y='release_year')
plt.title('Release Year Distribution by Type')
plt.show()
```
